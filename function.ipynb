```python
import numpy as np
from scipy.special import mathieu_a, mathieu_b, mathieu_cem, mathieu_sem
import matplotlib.pyplot as plt

def mathieu_equation_solution(a, q, x_range):
    n = 0  # Order of the solution
    a_n = mathieu_a(n, q)
    b_n = mathieu_b(n, q)

    x_values = np.linspace(x_range[0], x_range[1], 1000)
    ce_values, _ = mathieu_cem(n, q, x_values)
    se_values, _ = mathieu_sem(n, q, x_values)

    y1 = np.cos(a_n * x_values) * ce_values
    y2 = np.sin(b_n * x_values) * se_values

    return x_values, y1, y2

# Define the parameters a and q
a = 5
q = 2

# Define the x range for the plot
x_range = (0, 2 * np.pi)

# Solve the Mathieu equation
x_values, y1, y2 = mathieu_equation_solution(a, q, x_range)

# Plot the solutions
plt.figure(figsize=(10, 5))
plt.plot(x_values, y1, label="Mathieu C-Function Solution")
plt.plot(x_values, y2, label="Mathieu S-Function Solution")
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Mathieu Equation Solutions (a={a}, q={q})")
plt.legend()
plt.grid()
plt.show()
```
