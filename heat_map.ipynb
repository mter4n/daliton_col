```python
import numpy as np
from scipy.special import mathieu_a, mathieu_b, mathieu_cem
import matplotlib.pyplot as plt

def mathieu_equation_solution(a, q, x_range):
    n = 0  # Order of the solution
    a_n = mathieu_a(n, q)
    x_values = np.linspace(x_range[0], x_range[1], 1000)
    ce_values, _ = mathieu_cem(n, q, x_values)
    y = np.cos(a_n * x_values) * ce_values
    return y

# Define the range of a and q
a_range = np.linspace(0, 10, 100)
q_range = np.linspace(0, 10, 100)

# Define the x range for the plot
x_range = (0, 2 * np.pi)

# Compute the solution for each combination of a and q
z = np.zeros((len(a_range), len(q_range)))

for i, a in enumerate(a_range):
    for j, q in enumerate(q_range):
        y = mathieu_equation_solution(a, q, x_range)
        z[i, j] = np.mean(y)

# Plot the heatmap
plt.figure(figsize=(10, 8))
plt.imshow(z, origin='lower', aspect='auto', cmap='viridis',
           extent=[q_range[0], q_range[-1], a_range[0], a_range[-1]])
plt.colorbar(label='Mean of Solution')
plt.xlabel('q')
plt.ylabel('a')
plt.title('Heatmap of Mathieu Equation Solution')
plt.show()
```
